```python
#Aluno: João Pedro de Rossi Tambasco Calazans
#Matrícula: 202010405

#Importando Libs que serão utilizadas:
from PIL import Image
#Importando lib para transformar valores float em INT
from math import ceil
import cv2
from google.colab.patches import cv2_imshow
#Usaremos dlib para o reconhecimento facial
import dlib
```

```python
#Carregando o detector de faces do dlib, não é obrigatória a utilização de um arquivo externo,
#pois já tem um detector padrão do próprio dlib
face_detector = dlib.get_frontal_face_detector()

#Carregando a imagem que utilizarei
imagem = cv2.imread('faces.jpg')

#Carrega o detector de marcos faciais para o dlib:
detector_pontos = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')

#Encontra cada face localizada na imagem carregada anteriormente:
facesDetectadas = face_detector(imagem, 1)

#Criamos um loop para pegar todas as faces encontradas do código acima
for face in facesDetectadas:
  pontosDetectados = detector_pontos(imagem,face)

  #Para cada ponto detectado acima, desenhamos circulos pequenos vermelhos
  for ponto in pontosDetectados.parts():
    cv2.circle(imagem, (ponto.x, ponto.y), 2, (0,0,255), 1)

  #Por ultimo, desenhamos o circulo (AZUL) localizado no centro da face:
  l,t,r,b = face.left(), face.top(), face.right(), face.bottom()
  cv2.circle(imagem, (ceil((l+r)/2),ceil((t+b)/2)), ceil(((r-l)/2)), (255,0,0), 2 )


#E então, demonstramos o resultado:
cv2_imshow(imagem)





```
